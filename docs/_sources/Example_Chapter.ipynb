```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Here is a simple learning problem.

```python
x = np.linspace(-10, 10)
e = (np.random.rand(len(x)) - .5) * 20

def squared(x):
    return x**2

a = 2
b = 0
c = 5
def f(x, a, b, c):
    return a * squared(x) + b * x + c

y = f(x, a, b, c) + e

plt.plot(x, y)
plt.title('What curve is this?');
```

Now we will find the curve using linear regression

```python
A = np.vstack([squared(x), x, np.ones(len(x))]).T
a_hat, b_hat, c_hat = np.linalg.lstsq(A, y, rcond=None)[0]

print('Recovered quadratic:', a_hat, 'x^2 + ', a_hat, 'x + ', a_hat)
print('Original quadratic: 2 x^2 + 5')
```

```python
plt.plot(x, y)
plt.plot(x, f(x, a, b, c), label='Original Function')
plt.plot(x, f(x, a_hat, b_hat, c_hat), label='Derived Function')
plt.legend()
plt.title('Lets compare');
```
